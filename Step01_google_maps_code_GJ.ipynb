```python
#Now lets loop each request from our dataframe to append the contact number, info and website info:
def appending_num_web (row):
    row["Phone Number"] = np.nan
    row["International Phone Number"] = np.nan
    row["Website"] = np.nan
    row["Google maps URL"] = np.nan

    for number in range(len(row)): 
        #For future we need to change location for the location of the address which is in the end
        #Make google search nearby places with given specs:
        query_result = google_places.nearby_search(
        location='Lisbon', keyword= row["name"][number],
        radius=20000, language = "pt")
        time.sleep(random.random()+1)
        

        for place in query_result.places:
    # The following method has to make a further API call.
            place.get_details()
    # Referencing any of the attributes below, prior to making a call to
    # get_details() will raise a googleplaces.GooglePlacesAttributeError.
 #   print(place.details) # A dict matching the JSON response from Google.
            row["Phone Number"][number] = place.local_phone_number
            row["International Phone Number"][number] = place.international_phone_number
            row["Website"][number] = place.website
            row["Google maps URL"][number] = place.url

    return row 

appending_num_web(business_df)
```
